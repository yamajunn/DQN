```python
import cv2
import numpy as np
import pyautogui

# 画面の幅と高さを取得
screen_width, screen_height = pyautogui.size()

image_X = 320
image_Y = 180

# Qテーブルの初期化
num_states = image_X * image_Y
num_actions = 4  # 上下左右の移動
q_table = np.zeros((num_states, num_actions))

# ハイパーパラメータ
epsilon = 1.0  # ε-greedy法のε
epsilon_decay = 0.99  # εの減衰率
gamma = 0.99  # 割引率
learning_rate = 0.1  # 学習率

# 画像サイズと画面サイズの比率を計算
ratio_x = image_X / screen_width
ratio_y = image_Y / screen_height

# メインループ
while True:
    # 画面をキャプチャ
    screenshot = np.array(pyautogui.screenshot())

    # 画像をリサイズしてグレースケールに変換
    resized_image = cv2.resize(cv2.cvtColor(screenshot, cv2.COLOR_RGB2GRAY), (image_X, image_Y))

    # マウスカーソルの座標を取得
    x, y = pyautogui.position()

    # マウスカーソルの座標を画像の座標系に変換
    x_image = int(x * ratio_x)
    y_image = int(y * ratio_y)

    # 状態として使用する画像を準備
    state = resized_image.flatten()

    # ε-greedy法に基づいて行動を選択
    if np.random.rand() < epsilon:
        action = np.random.randint(num_actions)
    else:
        action = np.argmax(q_table[state])

    # 行動を実行し、報酬を取得
    if action == 0:  # 上
        pyautogui.move(0, -10)
    elif action == 1:  # 下
        pyautogui.move(0, 10)
    elif action == 2:  # 左
        pyautogui.move(-10, 0)
    elif action == 3:  # 右
        pyautogui.move(10, 0)

    # 新しい画像をキャプチャ
    new_screenshot = np.array(pyautogui.screenshot())
    new_resized_image = cv2.resize(cv2.cvtColor(new_screenshot, cv2.COLOR_RGB2GRAY), (image_X, image_Y))
    new_state = new_resized_image.flatten()

    # Q値を更新
    q_table[state][action] += learning_rate * (reward + gamma * np.max(q_table[new_state]) - q_table[state][action])

    # εを減衰
    epsilon *= epsilon_decay

    # 状態を更新
    state = new_state

    # 画面を表示
    cv2.imshow("Screen", resized_image)

    # キー入力を待機
    if cv2.waitKey(1) == ord("q"):
        break

cv2.destroyAllWindows()
```
